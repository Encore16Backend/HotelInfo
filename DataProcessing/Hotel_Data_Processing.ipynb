```python
import pandas as pd
import numpy as np
file_path = '/content/drive/MyDrive/data.xlsx'
df = pd.read_excel(file_path)
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['상세정보'].replace(' ', np.nan, inplace=True)
df.dropna(axis=0, subset=['상세정보'], inplace=True)
for c in df.columns:
    if df[c].isnull().sum() > 350:
        df.drop(c, axis=1, inplace=True)
df.drop('규모', axis=1, inplace=True)
df.drop('객실 수', axis=1, inplace=True)
df.drop('예약 안내', axis=1, inplace=True)
df.drop('객실 유형', axis=1, inplace=True)
```

```python
df.shape
```

```python
df.columns
```

```python
df.index=list(i for i in range(df['명칭'].count()))
```

```python
image = []
for i in df['명칭']:
    image.append("..\\assets\\images\\" + i + "1.jpg")
image = pd.Series(image)
print(image[:10])
print(image.shape)
```

```python
df.insert(26, 'hotelImage', image)
```

```python
df.head(1)
```

```python
images = []
for i in df['명칭']:
    tmp = ""
    for j in range(1, 4):
        if j != 3:
            tmp += "..\\assets\\images\\"+i+str(j)+'.jpg,'
        else:
            tmp += "..\\assets\\images\\"+i+str(j)+'.jpg'
    images.append(tmp)
print(images[:3])
images = pd.Series(images)
```

```python
df.insert(27, 'hotelImages', images)
```

```python
for i in range(df['명칭'].count()):
    if '[' in df.loc[i, '명칭']:
        name = df.loc[i, '명칭']
        df.loc[i, '명칭'] = name[:name.index('[')]
```

```python
for i in range(df['주소'].count()):
    if "\n" in df.loc[i, '주소']:
        address = df.loc[i, '주소']
        idx = address.index("\n")
        df.loc[i, '주소'] = address[:idx] 
```
